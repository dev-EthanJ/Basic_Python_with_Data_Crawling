# 알라딘 크롤링 with Python
# Aladin crawling with Python

```python
# crwaling library import

from bs4 import BeautifulSoup
from selenium import webdriver
import requests
# 코드 진행 지연을 위한 time 임포트

import time
# 2022-07 이후 selenium 업데이트로 인한 XPATH 추적 시 사용하는 임포트

from selenium.webdriver.common.by import By

# file io
import codecs
```

```python
my_driver = webdriver.Chrome()

# 알라딘 베스트셀러 사이트 접속
my_driver.get("https://www.aladin.co.kr/shop/common/wbest.aspx?BranchType=1&start=we")

# ip block 방지
time.sleep(10)

# 소스 얻기
my_source = my_driver.page_source

# parsing(page 전체)
my_parsed_source = BeautifulSoup(my_source, "html.parser")

my_driver.close()
```

```python
# web page element 조사 결과 > "li" tag에 책 정보(가격) 존재
# extract data by li tag
li_tag_data = my_parsed_source.find_all("li")

print(li_tag_data)
# 정확한 데이터 추출 실패
```

```python
# 책 정보를 감싸고 있는 "div" tag and "ss_book_box" class로 data extraction
div_book_box = my_parsed_source.find_all("div", class_="ss_book_box")

print(div_book_box)
print(len(div_book_box))
```

```python
# div_book_box data == bs4.element.ResultSet object > find_all() method 없음

print(type(div_book_box))
# div_book_box.find_all("li") > AttributeError
```

#### 내부 item 접근 방법
1. indexing으로 object 접근
2. loop 내부에서 하나하나 접근해서 확인

```python
print(type(div_book_box[0])) # > <class 'bs4.element.Tag'>

print(div_book_box[0]) # > 1위 책에 대한 정보 담긴 bs4.element.Tag object > .find_all() method 사용 가능
```

```python
# div_book_box[0] == first ranked book, "li" tag로 책 정보 찾기
first_li_book = div_book_box[0].find_all("li")

# extracted data item > text화
for item in first_li_book:
    print(item.text)
```

```python
# first_li_book[indexing].text > index 1 == title, index 2 == author, lindex 3 == price 추측됨 > print()해서 확인
print(first_li_book[1].text)
print(first_li_book[2].text)
print(first_li_book[3].text)
```

`div_book_box[index N == N+1순위 책].find_all("li")` > indexing by 1, 2, 3 > .text > 책 정보 출력 가능으로 추측됨   
\> index N에 다른 숫자 넣어서 확인

```python
for i in range(10, 14): # N = 10 ~ N = 13
    this_li_book = div_book_box[i].find_all("li")
    
    for j in range(1, 4): #index 1, 2, 3
        print(this_li_book[j].text)
        
    print()
```

\> 규칙성 확인 완료 > data crawling 가능!

### parsed_source를 이중 loop로 한 권씩 책 정보 출력   

- text.split() > 원하는 정보 골라서 출력 가능

```python
div_book_box = my_parsed_source.find_all("div", class_="ss_book_box")

title_list = list()
author_list = list()
price_list = list()

for this_div_book in div_book_box:
    this_li_book = this_div_book.find_all("li")
    
    this_title= this_li_book[1].text.split("]")[1]
    this_author = this_li_book[2].text.split(" (")[0]
    this_price = this_li_book[3].text.split("원 →")[0].replace(",", "，")
    
    print(this_title, this_author, this_price)
    
    title_list.append(this_title)
    author_list.append(this_author)
    price_list.append(this_price)
    
book_list = [title_list, author_list, price_list]
```

```python
for info_list in book_list:
    for info in info_list:
        print(info)
    
    print()
```
